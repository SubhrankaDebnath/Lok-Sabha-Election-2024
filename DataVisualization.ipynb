```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = r'C:\Users\BABIN\Desktop\Babin\Kalvium\loksava2024.csv'
df = pd.read_csv(file_path)

# Convert relevant columns to numeric
df['Total Votes'] = pd.to_numeric(df['Total Votes'], errors='coerce')
df['% of Votes'] = pd.to_numeric(df['% of Votes'], errors='coerce')

# Filter the dataset for independent and other candidates
independent_candidates = df[df['Party'] == 'Independent']
other_candidates = df[df['Party'] != 'Independent']

# Aggregate total votes and average percentage of votes for both groups
total_votes_independent = independent_candidates['Total Votes'].sum()
average_percentage_votes_independent = independent_candidates['% of Votes'].mean()

total_votes_others = other_candidates['Total Votes'].sum()
average_percentage_votes_others = other_candidates['% of Votes'].mean()

# Plotting total votes comparison
plt.figure(figsize=(10, 6))
plt.bar(['Independent', 'Others'], [total_votes_independent, total_votes_others], color=['orange', 'blue'])
plt.title('Total Votes: Independent vs Others')
plt.xlabel('Candidate Type')
plt.ylabel('Total Votes')
plt.show()

# Plotting average percentage of votes comparison
plt.figure(figsize=(10, 6))
plt.bar(['Independent', 'Others'], [average_percentage_votes_independent, average_percentage_votes_others], color=['orange', 'blue'])
plt.title('Average Percentage of Votes: Independent vs Others')
plt.xlabel('Candidate Type')
plt.ylabel('Average % of Votes')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r'C:\Users\BABIN\Desktop\Babin\Kalvium\loksava2024.csv'
df = pd.read_csv(file_path)

# Convert relevant columns to numeric
df['Total Votes'] = pd.to_numeric(df['Total Votes'], errors='coerce')

# Calculate the margin of victory for each constituency
# First, sort by constituency and total votes
df_sorted = df.sort_values(by=['constituency_name', 'Total Votes'], ascending=[True, False])

# Get the top two candidates for each constituency
top_two_candidates = df_sorted.groupby('constituency_name').head(2)

# Calculate the margin of victory
top_two_candidates['Rank'] = top_two_candidates.groupby('constituency_name')['Total Votes'].rank(ascending=False)
winning_margins = top_two_candidates.pivot(index='constituency_name', columns='Rank', values='Total Votes')
winning_margins['Margin'] = winning_margins[1] - winning_margins[2]

# Get the top 5 constituencies with the smallest winning margins
closest_margins = winning_margins.nsmallest(5, 'Margin')

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x=closest_margins['Margin'], y=closest_margins.index, palette='coolwarm')
plt.title('Top 5 Constituencies with Closest Winning Margins')
plt.xlabel('Winning Margin (Votes)')
plt.ylabel('Constituency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file
file_path = r'C:\Users\BABIN\Desktop\Babin\Kalvium\loksava2024.csv'  # Update with your file path
df = pd.read_csv(file_path)

# Step 2: Convert 'EVM Votes' and 'Postal Votes' to numeric
df['EVM Votes'] = pd.to_numeric(df['EVM Votes'], errors='coerce')
df['Postal Votes'] = pd.to_numeric(df['Postal Votes'], errors='coerce')

# Step 3: Calculate voter turnout percentage for each constituency
df['Total Votes'] = df['EVM Votes'].fillna(0) + df['Postal Votes'].fillna(0)
df['Voter Turnout Percentage'] = (df['Total Votes'] / df['Total Votes'].sum()) * 100

# Step 4: Calculate average voter turnout percentage per state
avg_turnout_per_state = df.groupby('state')['Voter Turnout Percentage'].mean().sort_values(ascending=False)

# Step 5: Plotting a bar chart
plt.figure(figsize=(12, 8))
avg_turnout_per_state.plot(kind='bar', color='orange')

plt.title('Average Voter Turnout per State')
plt.xlabel('State')
plt.ylabel('Average Voter Turnout Percentage (%)')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file
file_path = r'C:\Users\BABIN\Desktop\Babin\Kalvium\loksava2024.csv'  # Update with your file path
df = pd.read_csv(file_path)

# Step 2: Convert 'EVM Votes' and 'Postal Votes' to numeric
df['EVM Votes'] = pd.to_numeric(df['EVM Votes'], errors='coerce')
df['Postal Votes'] = pd.to_numeric(df['Postal Votes'], errors='coerce')

# Step 3: Calculate total votes and voter turnout percentage for each constituency
df['Total Votes'] = df['EVM Votes'].fillna(0) + df['Postal Votes'].fillna(0)
df['Voter Turnout Percentage'] = (df['Total Votes'] / df['Total Votes'].sum()) * 100

# Step 4: Sort by voter turnout percentage in ascending order
df_sorted = df.sort_values(by='Voter Turnout Percentage')

# Step 5: Select the top 10 constituencies with the least voter turnout percentage
smallest_6_constituencies = df_sorted.head(10)[['constituency_name', 'Voter Turnout Percentage']]

# Step 6: Plotting a bar chart
plt.figure(figsize=(12, 8))
plt.bar(smallest_6_constituencies['constituency_name'], smallest_6_constituencies['Voter Turnout Percentage'], color='skyblue')

plt.title('Top 6 Constituencies with the Least Voter Turnout')
plt.xlabel('Constituency Name')
plt.ylabel('Voter Turnout Percentage (%)')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```
